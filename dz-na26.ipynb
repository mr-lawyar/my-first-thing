Задача №2

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.loadtxt("chile_mag.txt")
magn = data    # магнітуда
```

```python
at_least_5 = np.sum(magn >= 5)
at_least_6 = np.sum(magn >= 6)
at_least_7 = np.sum(magn >= 7)

print(at_least_5)
print(at_least_6)
print(at_least_7)

print("   ") # просто пропуск

print(at_least_5 / at_least_6)
print(at_least_6 / at_least_7) # нічого не помітив особливого :(
```
